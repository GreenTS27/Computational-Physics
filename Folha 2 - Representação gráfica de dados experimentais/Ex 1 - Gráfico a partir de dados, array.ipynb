```python
from numpy import loadtxt
from pylab import plot,show

data = loadtxt(r"C:\Users\tiago\Documents\FCUP\2º ano (20-21) - 2º semestre\Física Computacional\Exercícios\Programas e dados - Mark Newman\sunspots.txt",float)
# Se o ficheiro estivesse na msm pasta -->  data = loadtxt("sunspots.txt")

x = data[: ,0]
y = data[: ,1]

plot(x,y)
show()
```

```python
x_1000 = data[: 1000,0]
y_1000 = data[: 1000,1]

plot(x_1000,y_1000)
show()
```

```python
r = 5
def MedManchas(yk): # média dos valores de manchas no intervalo [yk-r , yk+r]
    
    return 1 / (2*r + 1) * sum(yk)
    
Y = []
for i in range(1000):
    Y.append( MedManchas(y_1000[i-r : i+r+1]) )
    
plot(x_1000,y_1000)
plot(x_1000,Y)
show()
```
