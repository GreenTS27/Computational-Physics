```python
import matplotlib.pyplot as plt
import numpy as np


teta = np.linspace(0, 2*np.pi)
x = 2 * np.cos(teta) + np.cos(2*teta)
y = 2 * np.sin(teta) - np.sin(2*teta)
plt.plot(x,y)
plt.show()
```

```python
def EspiralGalileu(teta_limite, teta_steps): # quanto maior for teta_steps,maior ser√° a curvatura
    teta = np.linspace(0, teta_limite, teta_steps)
    r = teta ** 2
    return r, teta

def CoordenadasPolares(r, teta):
    x = r * np.cos(teta)
    y = r * np.sin(teta)
    return x, y

R,Teta = EspiralGalileu(10*np.pi, 100)
X,Y = CoordenadasPolares(R,Teta) 

plt.plot(X,Y)
```

```python
def Fey(teta_lim, teta_step):
    teta = np.linspace(0, teta_lim, teta_step)
    r = np.exp( np.cos(teta) ) - 2*np.cos(4*teta) + ( np.sin(teta/12) )**5
    return r, teta

R,Teta = Fey(24*np.pi, 1000)
X,Y = CoordenadasPolares(R,Teta)
plt.plot(X,Y)
```
