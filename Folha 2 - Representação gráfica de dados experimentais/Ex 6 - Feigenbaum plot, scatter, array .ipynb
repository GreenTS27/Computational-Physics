```python
import numpy as np
import matplotlib.pyplot as plt

def caos(r):
    x = 1/2
    for i in range(1001): # se variarmos o nº de iteraçoes os ramos variam 
        x = r * x * (1 - x)
    return x

r = np.arange(1, 4, 0.01)
for j in r:
    #print(j,caos(j))
    #print("---")
    plt.scatter(j,caos(j),s=1,color="black")
plt.ylabel("x")
plt.xlabel("r")
plt.show()
#um x para cada r
```
