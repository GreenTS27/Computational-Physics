```python
import numpy as np
import matplotlib.pyplot as plt
import time 

def Mandelbrot(limites,N,num_iterar):
    x = np.linspace(-limites,limites, N, dtype=np.float64) #matriz NxN
    y = x
    c = x + 1j * y[:,None] #c = x + i*y

    z = np.zeros(c.shape,dtype=np.complex64) #matriz de zeros com a msm forma de c
    grelha = np.zeros(c.shape)
    
    for i in range(num_iterar):
        poligrafo = np.less_equal(np.sqrt(z.real**2 + z.imag**2), 2)
        z[poligrafo] = z[poligrafo]**2 + c[poligrafo] #usamos apenas os valores de z em que |z|<2
        
        for k in range(N):
            for l in range(N):
                if poligrafo[k,l] == True: #está na série Mandelbrot
                    grelha[k,l] = i
                    
    plt.figure(figsize = (7,7))
    plt.imshow(grelha,extent=[-limites, limites, -limites, limites])
    plt.ylabel("y")
    plt.xlabel("x")
    plt.copper()
    plt.show()
```

```python
%%time
Mandelbrot(1,450,150)
```
