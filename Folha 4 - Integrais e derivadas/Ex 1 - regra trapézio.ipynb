```python
import numpy as np
import matplotlib.pyplot as plt

dados = np.loadtxt(r"C:\Users\tiago\Documents\FCUP\2º. ano (2020 - 2021)\2 - 2º semestre\Física Computacional\Programas e dados - Mark Newman\velocities.txt")

tempo = dados[:,0] #eixo x
vel = dados[:,1]   #eixo y
N = len(tempo)

def x(t): # dist percorrida pela particula no instante t
    if t == 0: 
        return 0
    a = tempo[0]
    b = tempo[int(t)] # um índice é sempre um nº inteiro
    h = (b - a) / t
    return h * ( 0.5*vel[int(a)] + 0.5*vel[int(b)] + sum( vel[1: int(t)] ) ) # ver equação 5.3 do livro

dist = np.zeros(N)
for i in range( N ):
    dist[i] = x(i)

plt.plot(tempo, vel)
plt.plot(tempo, dist)
plt.legend(("Velocidades", "Distância percorrida"))
plt.xlabel("t (s)")
plt.show()

##############################################################################
deslocamento = 0
for k in range(N):
    deslocamento += x(k)
print("dist. total percorrida/deslocamento =", deslocamento)
```
