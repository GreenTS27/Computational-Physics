#### Derivada ordinária

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return 1 + 0.5 * np.tanh(2*x) 

def d_difCentral(x,f):
    h = 10 ** -5
    return (f(x + h/2) - f(x - h/2)) / h

def g(x): # derivada analítica
    return 1 / np.cosh(2*x)**2

N = 100

x = np.linspace(-2,2,N)
d_pc = np.zeros(x.shape)
for i in range(N):
    d_pc[i] = d_difCentral(x[i], f)

d_analitica = list(map(g, x))

plt.plot(x,d_pc,"go")
plt.plot(x,d_analitica,"k")
plt.xlabel('x')
plt.title(r"$\sech^2(x)$")
plt.show()
```

#### Derivada parcial

```python
def dParcial(f, x, y): # derivada parcial, 2 variáveis
    #método diferença central
    h = 10 ** -5
    df_dx = (f(x + h/2 , y) - f(x - h/2 , y)) / h
    df_dy = (f(x , y + h/2) - f(x , y - h/2)) / h
    return df_dx , df_dy
```
