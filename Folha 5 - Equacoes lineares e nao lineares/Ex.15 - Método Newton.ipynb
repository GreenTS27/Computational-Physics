```python
import numpy as np
import matplotlib.pyplot as plt

def P(x):
    return 924*x**6 - 2772*x**5 + 3150*x**4 - 1680*x**3 + 420*x**2 - 42*x + 1

nPontos = 100
x = np.linspace(0.0,1.0, nPontos)
yP = np.zeros(x.shape)
y0 = np.zeros(x.shape)
for i in range(nPontos):
    yP[i] = P(x[i])
    
plt.plot(x, yP, "g")
plt.plot(x, y0, "k")
plt.show()    

def d_P(x):
    return 6*924*x**5 - 5*2772*x**4 + 4*3150*x**3 - 3*1680*x**2 + 2*420*x - 42


def RaizNewton(funç, d_funç, x, precisão):
    delta = 1.0 #isto é o erro
    while abs(delta) > precisão:
        delta = funç(x) / d_funç(x)
        x -= delta
    return x


print('As seis raízes de P(x) são:')
print(RaizNewton(P, d_P, 0, 10**-12))
print(RaizNewton(P, d_P, 0.2, 10**-12))
print(RaizNewton(P, d_P, 0.4, 10**-12))
print(RaizNewton(P, d_P, 0.6, 10**-12))
print(RaizNewton(P, d_P, 0.8, 10**-12))
print(RaizNewton(P, d_P, 0.95, 10**-12))
```
