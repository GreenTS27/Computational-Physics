```python
# Vem do ficheiro de Newman dft.py

from numpy import zeros,loadtxt
from pylab import plot,xlim,show
from cmath import exp,pi

def dft(y):
    N = len(y)
    c = zeros(N//2+1,complex)
    for k in range(N//2+1):
        for n in range(N):
            c[k] += y[n]*exp(-2j*pi*k*n/N)
    return c

y = loadtxt("pitch.txt",float)
c = dft(y)
plot(abs(c))
xlim(0,500)
show()
```

```python
import numpy as np
from numpy import exp, arange, pi, zeros,linspace, ones
from pylab import *

def dft(y):
    N = len(y)
    c = zeros(N//2+1,complex)
    for k in range(N//2+1):
        for n in range(N):
            c[k] += y[n]*exp(-2j*pi*k*n/N)
    return c


N = 1000

print("Onda quadrarda")
y1 = ones(N)
y1[:N//2]=0
plot(y1)
show()
c1 = dft(y1)
plot(abs(c1[200:300])) # <-----------
show()

print("Onda dente de serra")
y2 = arange(N)
plot(y2)
show()
c2 = dft(y2)
plot(abs(c2))
show()

print("Onda sinusoidal")
n = arange(N)
y3 = sin(pi*n/N)*sin(20*pi*n/N)
plot(y3)
show()
c3 = dft(y3)
plot(abs(c3))
show()
```
