```python
from scipy.fftpack import fft
from scipy import loadtxt
from pylab import plot, show, xlabel, ylabel

print("----------------------------------- Piano ----------------------------------")
piano = loadtxt('piano.txt', float)
plot(piano)
show()

piano_fourier = fft(piano)
plot(abs(piano_fourier))
show()
plot(abs(piano_fourier[0:10000]))
xlabel('k')
ylabel('$|c_k|$')
show()

print("------------------------- Trompete é melhor que piano ----------------------")

trumpet = loadtxt('trumpet.txt', float)
plot(piano)
show()

trumpet_fourier = fft(trumpet)
plot(abs(trumpet_fourier))
show()
plot(abs(trumpet_fourier[0:10000]))
xlabel('k')
ylabel('$|c_k|$')
show()

# the frequency associated with k is given by f =  k * (44100 samples/s) / 100000 samples
```
