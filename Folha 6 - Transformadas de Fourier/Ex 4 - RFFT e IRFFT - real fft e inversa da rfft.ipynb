```python
from scipy import loadtxt, zeros, copy, floor
from scipy.fftpack import rfft, irfft
from pylab import plot, show, xlabel, ylabel,legend

dow = loadtxt("dow.txt", float)
# plot(dow)
# show()

dow_fourier = rfft(dow)
N = len(dow_fourier)
print(N)

#retem 10% dos valores de dow
first_10_percent = zeros(N, float)
first_10_percent[: N // 100] = copy(dow_fourier[ : N // 100])
dow_first_10_p = irfft(first_10_percent)

#retem 2% dos valores de dow
first_2_percent = zeros(N, float)
first_2_percent[ : N // 50] = copy(dow_fourier[ : N // 50])
dow_first_2_p =irfft(first_2_percent)

plot(dow, 'k', label= "original")
plot(dow_first_10_p, 'r', label = "remove 90%")
plot(dow_first_2_p, 'g', label="remove 98%")
legend()
show()
```

```python
from pylab import *
from numpy import loadtxt
from numpy.fft import rfft, irfft

data = loadtxt('dow.txt')
N = data.shape[0]
plot(data,"k",label='original')

c1 = rfft(data)
c1[N//10:]=0
y1=irfft(c1)

c2 = rfft(data)
c2[N//50:]=0
y2=irfft(c2)

c3 = rfft(data)
c3[N//100:]=0
y3=irfft(c3)

plot(y1,label=' remove 90%')
plot(y2,label=' remove 98%')
plot(y3,label=' remove 99%')
legend()
show()
```
