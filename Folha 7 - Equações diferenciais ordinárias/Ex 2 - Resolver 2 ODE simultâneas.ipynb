$x = $ população de coelhos em milhares 

$y = $ população de raposas em milhares

Variação do número de coelhos:
$$f_x = \frac{dx}{dt} = \alpha x - \beta xy$$

Variação do número de raposas:
$$f_y = \frac{dy}{dt} = \gamma xy - \delta y$$

Sabemos que $\alpha$, $\beta$, $\delta$ e $\gamma$ são constantes.

```python
import numpy as np
import matplotlib.pyplot as plt

alfa = 1
beta = 0.5
gama = 0.5
delta = 2
x0 = 2
y0 = 2

r = np.array([x0,y0],float) 
def f(r,t):
    x = r[0]
    y = r[1]
    fx = alfa*x - beta*x*y
    fy = gama*x*y - delta*y
    return np.array([fx,fy] ,float)

a = 0
b = 30
N = 1000
h = (b-a) / N

tPontos = np.arange(a,b,h)
xPontos = []
yPontos = []

for t in tPontos:
    xPontos.append(r[0])
    yPontos.append(r[1])
    k1 = h*f(r,t)
    k2 = h*f(r+0.5*k1,t+0.5*h)
    k3 = h*f(r+0.5*k2,t+0.5*h)
    k4 = h*f(r+k3,t+h)
    r += (k1 + 2*k2 + 2*k3 + k4)/6    

plt.figure(figsize = (10,5))
plt.plot(tPontos,xPontos,label = "Coelhos ")
plt.plot(tPontos,yPontos,label = "Raposas ")
plt.legend()
plt.xlabel("Tempo (em anos?)")
plt.ylabel("População em milhares")
plt.show()

# tPontos2 = np.linspace(a,b-h,N)
# print(tPontos2)
```

Quado há muitos coelhos a população de raposas aumenta até chegar um ponto em que há demasiadas raposas a caçar coelhas, pelo que a população de coelhos diminui.

Se a população de coelhos diminui então a população de raposas também vai diminuir, pois estas não têm comida suficiente para sobreviver.

Contudo, chega uma certa altura em que há tão poucas raposas que a população de coelhos consegue voltar a aumentar e, assim, repete-se o ciclo.
