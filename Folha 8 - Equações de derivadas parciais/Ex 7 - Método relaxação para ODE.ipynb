$$ \frac{d^2x}{dt^2} = -g $$

$$ f''(x) = \frac{f(x+h) - 2f(x) + f(x-h)}{h^2} $$

$$ f(x) = \frac{1}{2} \left[f''(x) h^2 - f(x+h)  - f(x-h) \right]  $$

$$ f(x) = \frac{1}{2} \left[g h^2 + f(x+h) + f(x-h) \right]  $$

```python
import numpy as np
import matplotlib.pylab as plt

g = 9.81 # m/ (s**2)
N = 100 # pontos da grelha
t0 = 0
tf = 10
h = (tf - t0)/N
precisao = 10**-6


x2 = np.zeros(N+1,float) # valor inicial 
erro = 2*precisao
while erro > precisao: 
    erro = 0.0
    for i in range(N):
        x1 = x2[i]
        x2[i] = 1/2 * (g*h**2 + x2[i - 1] + x2[i + 1]) #f(x1)
        erro = max(erro, abs(x1 - x2[i]))

valores_t = np.linspace(t0,tf,N+1)
plt.plot(valores_t , x2)
plt.xlabel("t (s)")
plt.ylabel("y (m)")
plt.show()
```
